### BASIC INFO & CLEANUP

```python
# Check data

print(df.info())

print(df.describe())

# Remove missing Customer IDs

df. dropna (subset=['CustomerID'])

# Remove negative quantities

df = df [df['Quantity'] > 0]

# Convert InvoiceDate to datetime

df ['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create TotalPrice column

df['TotalPrice'] = df['Quantity'] * df ['UnitPrice']
```

### PRODUCT ANALYSIS-REVENUE-GENERATING PRODUCTS

```python
top_revenue_products = (df.groupby('Description')['TotalPrice']

.sum()

.sort_values (ascending=False)

.head(10)

)

print(top_revenue_products)
```

```python
top_revenue_products.plot(kind='bar',figsize=(10,6))

plt.title("Top 10 Products by Revenue")

plt.ylabel("Revenue")

plt.xlabel("Product Description")

plt.xticks(rotation=45)

plt.show()
```

Know you read to visualize the top 10 products based on revenue.
It creates a bar chart using the plot() function.
The bar chart helps compare which products generate the highest sales.
figsize=(10,6) defines the size of the chart for better visibility.
The chart title "Top 10 Products by Revenue" explains what the chart represents.
The Y-axis is labeled as Revenue, showing total sales amount.
The X-axis is labeled as Product Description, showing product names.
X-axis labels are rotated by 45 degrees to avoid overlapping and improve readability.
plt.show() displays the final chart.
This visualization helps in business decision-making, such as identifying top-selling products

### MONTHLY REVENUE TREND

```python
df ['Month'] = df ['InvoiceDate'].dt.to_period('M')

monthly_revenue = (df.groupby('Month') ['TotalPrice'] .sum()

                  )

monthly_revenue.plot(kind='line',

figsize=(12,6))

plt.title("Monthly Revenue Trend")

plt.ylabel("Revenue")

plt.xlabel("Month")

plt.show()
```

Converts the date into a monthly period format.

Calculates total revenue per month by summing TotalPrice.
